# boosting

## Adaboosting

adaboosting是boosting族的一个算法，用于二分类

参数：训练基础学习器个数T  
所有样本赋值权重一样1/m，训练一个基础的学习器，然后根据基础学习器的分类误差进行对训练样本权重更新  
学习器的权重和其在训练样本上的分类误差有关，1/2\*ln(1-error/error)  
直到基础学习器的个数到达T，使用线性的方式求和  

## GBDT(gradient boosting decision tree)

[GBDT](https://www.cnblogs.com/pinard/p/6140514.html)  
使用泰勒一阶展开，梯度下降法

# bagging(bootstrap aggregating)

booststrap sampling：有放回采样，m个样本采m次，则当m很大时，大概有63.2%的训练样本被采集  
进行多次采样，训练多个学习器

## 随机森林(random forest)RF

若训练样本有n个属性，从中选k个属性做组合，训练一系列的决策树  
k = log2(n)

基模型使用CART(Classification And Regression Tree)

# boosting VS bagging

- 一个串行，一个并行
- 一个关注低偏差，一个关注低方差
- 一个有权重，一个没有权重
- 一个分类器强相关，一个分类器之间弱相关

通俗点的说法：哈哈～～～
一个是随机bagging，一个是boosting。bagging就是大家都是学渣，每道题都由随机选出的一群学渣投票决定，这样需要的学渣比较多，而且每个学渣还都得很努力学习。boosting也是一群学渣，但每个人虽然总分菜，却是因为偏科导致的，每个学渣都贡献自己最擅长的那个题目。这样boosting需要的每个学渣都豪不费力，但是整体上更强了。xgb的学渣还通过预习，让自己偏科的科目学得更省力。所以整体上xgb看起来是非常省力的一群学渣组成，但是拿到的分数却很高。
