### 类别标签编码[sklearn.preprocessing.LabelEncoder](http://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.LabelEncoder.html#sklearn.preprocessing.LabelEncoder)

encodes labels with values between 0 and n_classes-1.

```python
le = preprocessing.LabelEncoder()

# 按照字典序进行编码
le.fit(['a', 'd', 'b', 'c', 'd'])
print(le.classes_)

print(le.transform(['a', 'd', 'b', 'c', 'd']))

print(le.inverse_transform([0, 1, 3, 0]))
```
